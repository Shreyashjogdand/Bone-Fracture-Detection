```python
def to_uppercase(column):
  return column.upper()
```

```python
data.loc[:,'Capitalized_title'] = data.loc[:,'title'].apply(to_uppercase)
```

```python
def upper_case(value):
  return value.upper()
```

```python
data.loc[:,'original_title']= data.loc[:,'original_title'].apply(upper_case)
```

```python
def lower_case(value):
  if isinstance(value , str):#there are misising values in tagline hence using isinstance
     return value.lower()
  else:
    return value
```

```python
data.loc[:,'tagline']=data.loc[:,'tagline'].apply(lower_case)
```

```python
# create profit column
def calculate_profit(row):
    return row['revenue'] - row['budget']
```

```python
data.loc[:,'profit']=data.apply(calculate_profit,axis=1)
data
```

```python
# Define the function to calculate revised profit
def revised_profit(revenue, budget):
    return revenue - budget  # Modify this logic if needed

# Apply revised_profit across rows to create a new 'new_profit' column
data['new_profit'] = data.apply(lambda x: revised_profit(x['revenue'], x['budget']), axis=1)
```

```python
# convert original_lan into upper case using lambda function
data.loc[:,'original_language']=data.loc[:,'original_language'].apply(lambda x : x.upper())
```

```python
# round popularity values to 2 decimal places
data.loc[:,'popularity_new']=data.loc[:,'popularity'].apply(lambda x : round(x , 2))
```

```python
# create new column profit_value by change changeing positive profit as "positive" and negative value should be change as "negative"
data.loc[:,"profit_value"]=data.loc[:,'profit'].apply(lambda x:"positive" if x >0 else "negative")
```

```python
def func(value, key):
    try:
        j = eval(value)  # Convert string to Python object
        df = pd.json_normalize(j)  # Flatten JSON into DataFrame
        names = list(df[key])  # Extract values of 'name' column
        return ",".join(names)  # Return as comma-separated string
    except:
        return value  # Return original value if error occurs
```

```python
data.loc[:,'genre_names']=data.loc[:,'genres'].apply(func,key='name')
data.head()
```
